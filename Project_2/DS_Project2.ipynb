```
from sklearn.model_selection import train_test_split 
import pandas as pd 
import numpy as np 
from sklearn import metrics 
import pandas as pd
import collections
import random

dataCredit = pd.read_csv("/content/drive/My Drive/datsets_datascience/creditcardfraud/creditcard.csv")
dataCredit
```

```
class_count = collections.Counter(dataCredit['Class'])
print(class_count)
print("Imbalance-->", (class_count[0]/(class_count[0] + class_count[1]))*100,'%')
#downsampling majority class
indexNames = dataCredit[(dataCredit['Class'] == 0)].index
PercentageToRemove = 0.995
ToRemove = int(class_count[0] * PercentageToRemove)
indexToRemove = random.sample(indexNames.tolist(),ToRemove)
dataCredit = dataCredit.drop(indexToRemove)

class_count = collections.Counter(dataCredit['Class'])
print(class_count)
print("Imbalance afer downsampling-->",(class_count[0]/(class_count[0] + class_count[1]))*100,'%' )
```

```
dataCredit['Class'].value_counts().head(2).plot(kind = 'bar')
```
